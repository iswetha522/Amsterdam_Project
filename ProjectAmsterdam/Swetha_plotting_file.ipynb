```python
import pandas as pd
import os
```

```python
votes_2002 = pd.read_csv('2002_jaarboek.csv', skiprows=0, sep=';', usecols=[0,1,2,3,5], na_filter=False)
```

```python
political_parties_2002 = pd.DataFrame()
political_parties_2002['Info'] = ['Total Votes', 'Percentage']
political_parties_2002['CDA'] = ['21460','7,9' ]
political_parties_2002['PVDA'] = ['77792','28,6' ]
political_parties_2002['VVD'] = ['48804','18,0' ]
political_parties_2002['GroenLinks'] = ['35197','13,0' ]
political_parties_2002['SP'] = ['20777','7,7' ]
political_parties_2002['D66'] = ['19600','7,2' ]
political_parties_2002 = political_parties_2002.transpose()
```

```python
votes_2006 = pd.read_csv('2006_gemeenteraad.csv', skiprows=0, usecols=[0,1,2,4,6], sep=';', na_filter=False)

# votes_2006
```

```python
political_parties_2006 = pd.DataFrame()
political_parties_2006['Info'] = ['Total Votes', 'Percentage']
political_parties_2006['CDA'] = ['12715','4,2' ]
political_parties_2006['PVDA'] = ['119651','39,4' ]
political_parties_2006['VVD'] = ['49172','16,2' ]
political_parties_2006['GroenLinks'] = ['41563','13,7' ]
political_parties_2006['SP'] = ['40637','13,4' ]
political_parties_2006['D66'] = ['12338','4,1' ]
political_parties_2006 = political_parties_2006.transpose()
```

```python


combined_2002_2006_df = political_parties_2002.join(political_parties_2006, lsuffix='_2002', rsuffix='_2006')
combined_2002_2006_df = combined_2002_2006_df.reset_index()
combined_2002_2006_df = combined_2002_2006_df.drop(0) 
combined_2002_2006_df
```

```python
import matplotlib.pyplot as plt
import pandas as pd

ax = plt.gca()

combined_2002_2006_df["0_2002"]= combined_2002_2006_df["0_2002"].astype(float)
combined_2002_2006_df["0_2006"]= combined_2002_2006_df["0_2006"].astype(float)

combined_2002_2006_df.plot(kind='line',x='index',y='0_2002',ax=ax)
combined_2002_2006_df.plot(kind='line',x='index',y='0_2006', color='red', ax=ax)

plt.show()
```
